<h1 style="font-weight: bold;">MULTIPLE LINEAR REGRESSION with SCIKIT-LEARN</h1>


<p style="font-size: 20px; font-weight: 400;"><i>The Dataset in CSV format</i> is converted to <i>a Pandas Dataframe</i>.</p>

```python
import pandas as pd

df = pd.read_csv("multiple_linear_regression_data.csv")
df.head(5)
```

<p style="font-size: 20px; font-weight: 400;"><i>The Dataframe</i> is divided into <i>Features (X)</i> and <i>Labels (y)</i> datasets.</p>

```python
X, y = df.iloc[:,:-1], df.iloc[:,-1]
```

<p style="font-size: 20px; font-weight: 400;"><i>The Features (X)</i> and <i>Labels (y)</i> are splitted into their corresponding <i>Train</i> and <i>Test</i> sets, with <i>the Test size</i> being <i>one-fifth</i> of the whole dataset.</p>

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

<h2 style="font-weight: bold;">Multiple Linear Regression</h2>
<p style="font-size: 20px; font-weight: 400;"><i>Multiple Linear Regression</i> is a statistical method used to model the <i>linear relationship</i> between <i>multiple independent variables (X<sub>1</sub>,X<sub>2</sub>,..., X<sub>m</sub>)</i> and <i>single dependent variable (Y)</i>. In this model, the dependent variable, is forecasted based on the independent variables.</p>
<br>
<p style="font-size: 20px; font-weight: 400;"><i>Formula: </i></p>
<div style="text-align: left;">
    <p style="text-align: center; font-size: 20px; font-weight: 400;"">y = &beta;<sub>0</sub> + &beta;<sub>1</sub> x<sub>1</sub> + &beta;<sub>2</sub> x<sub>2</sub> + ... + &beta;<sub>m</sub> x<sub>m</sub> + &epsilon;</p>
                  <p style="text-align: right; font-size: 16px;"><strong>y :</strong> Dependent Variable &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;</p>
             <p style="text-align: right; font-size: 16px;"><strong>x<sub>i</sub> :</strong> Independent Variables &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;</p>
         <p style="text-align: right; font-size: 16px;"><strong>m: </strong>Number of Independent Variable &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;</p>
     <p style="text-align: right; font-size: 16px;"><strong>&beta;<sub>0</sub>:</strong> Intercept Value&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;</p>
    <p style="text-align: right; font-size: 16px;"><strong>&beta;<sub>i</sub>:</strong> Coefficient Values&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;</p>
    <p style="text-align: right; font-size: 16px;"><strong>&epsilon;:</strong> Error Value &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;</p>
    
</div>

<p style="font-size: 20px; font-weight: 400;"><i>The Linear Regression Model Object</i> is defined by utilizing <i>Scikit-Learn</i> library. It is fitted using the <i>training features</i>. Thus, the model was successfully created.</p>

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)

type(lr)
```

<p style="font-size: 20px; font-weight: 400;">Then,<i> the Linear Regression Formula</i> is derived.</p>

```python
subChars = "₁₂"
equation = f"y = { round(lr.intercept_, 3) } + { round(lr.coef_[0], 3) } x{subChars[0]} + { round(lr.coef_[1], 3) } x{subChars[1]}"

print(f"The Multiple Linear Regression Formula : { equation }")
```

<p style="font-size: 20px; font-weight: 400;">The model is used to predict <i>the test labels</i> using <i>the test features</i> dataset.</p>

```python
preds = lr.predict(X_test)
preds
```

<h2 style="font-weight: bold;">Mean Squared Error</h2>
<p style="font-size: 20px; font-weight: 400;"><i>The Mean Squared Error</i> is the average squared difference between <i>observed</i> and <i>predicted values.</i> MSE is frequently used for evaluating the performance of regression models.</p>
<p style="font-size: 20px; font-weight: 400;"><i>Formula: </i></p>
<div style="text-align: left;">
    <p style="font-size: 20px; font-weight: 400;">$$\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \ p_i)^2$$</p>
    <p style="text-align: right; font-size: 16px;"><strong>n:</strong> Number of Instances (Rows)&nbsp;&nbsp;&nbsp;&nbsp;</p>
    <p style="text-align: right; font-size: 16px;"><strong>y<sub>i</sub>:</strong> \(i\)th Actual Y Value&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;</p>
    <p style="text-align: right; font-size: 16px;"><strong>p<sub>i</sub>:</strong> \(i\)th Prediction Value&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;</p>

</div>

<p style="font-size: 20px; font-weight: 400;"><i>The Mean Squared Error</i> is calculated by using <i>Predictions</i> and <i>Actual Y Values.</i></p>

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, preds)
print("The Mean Squared Error:", round(mse, 3))
```

<p style="font-size: 20px; font-weight: 400;">In the PairGrid chart below, the relationships between <i>Y</i> and <i>X<sub>1</sub></i>, as well as between <i>Y</i> and <i>X<sub>2</sub></i>, are demonstrated <i>individually</i>.</p>

```python
import seaborn as sns
import matplotlib.pyplot as plt

grid = sns.PairGrid(df, y_vars=df.columns[-1], x_vars=df.columns[:-1], height=4.4)
grid.map(sns.scatterplot)
plt.show()
```

<p style="font-size: 20px; font-weight: 400;">As it seen in the charts above, <i>Y</i> has a <i>negative</i> linear relationship with <i>X<sub>1</sub></i> while having a <i>positive</i> linear relationship with <i>X<sub>2</sub></i>.</p>

<p style="font-size: 20px; font-weight: 400;">The 3D graph below illustrates <i>the Actual Y Values</i> in the Test set and <i>the Flat Prediction Surface</i> created by Linear Regression, using grid values between 0 and 300 for X<sub>1</sub> and X<sub>2</sub>.</p>

```python
import warnings
import numpy as np
from mpl_toolkits.mplot3d import Axes3D 
warnings.filterwarnings("ignore", category=UserWarning)

fig = plt.figure(figsize=(7,7)) 
ax = Axes3D(fig, auto_add_to_figure=False) 

# Actual Y Values (DARK BLUE)
fig.add_axes(ax)
ax.scatter(X_test.iloc[:,0], 
           X_test.iloc[:,1], 
           y_test, 
           s=30, 
           marker='o', 
           alpha=1,
           color="darkblue")

# Grid
x1, x2 = np.meshgrid(np.linspace(0, 300, 300), np.linspace(0, 300, 300))
X_grid = np.hstack((x1.reshape(-1, 1), x2.reshape(-1, 1)))

# Predictions for Grid
preds = lr.predict(X_grid)


# Multiple Linear Regression Surface (COOL COLOR SCALE)
ax.plot_surface(x1, x2, preds.reshape(x1.shape), alpha=0.5, cmap='cool')

ax.set_xlabel("X1") 
ax.set_ylabel("X2") 
ax.set_zlabel("Y")  
plt.show()
```
